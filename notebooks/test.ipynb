```python
import threading

shared_var = 0
semaphore = threading.Semaphore(1)

def incrementer():
    global shared_var
    while True:
        semaphore.acquire()
        print("Hilo 1: valor de la variable compartida antes de incrementar = ", shared_var)
        shared_var += 1
        print("Hilo 1: valor de la variable compartida después de incrementar = ", shared_var)
        semaphore.release()
        if shared_var >= 2:
            break

def decrementer():
    global shared_var
    while True:
        semaphore.acquire()
        print("Hilo 2: valor de la variable compartida antes de decrementar = ", shared_var)
        shared_var -= 1
        shared_var //= 2
        print("Hilo 2: valor de la variable compartida después de decrementar y dividir por 2 = ", shared_var)
        semaphore.release()
        if shared_var <= -2:
            break

t1 = threading.Thread(target=incrementer)
t2 = threading.Thread(target=decrementer)

t1.start()
t2.start()

t1.join()
t2.join()
```
