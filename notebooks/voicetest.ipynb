```python
import pyttsx3

# Inicializar el motor de texto a voz
engine = pyttsx3.init()

# Configurar la voz
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id) # Cambia el número 0 por otro para elegir una voz diferente

# Cambiar la velocidad de habla
engine.setProperty('rate', 142) # Cambia el número 150 por otro para cambiar la velocidad

# Convertir texto a voz
texto = "Abriendo Youtube..."
engine.say(texto)
engine.runAndWait()
```

```python
from gtts import gTTS
from io import BytesIO
import simpleaudio as sa

# Convertir texto a voz
texto = "Hola, ¿cómo estás?"
tts = gTTS(text=texto, lang='es')
audio_file = BytesIO()
tts.write_to_fp(audio_file)

# Cargar audio desde memoria y reproducir
audio_file.seek(0)
audio_data = audio_file.read()
play_obj = sa.play_buffer(audio_data, 1, 2, 22050)
play_obj.wait_done()
```

```python
import pyttsx3

def speak(text, voice_id=None):
    engine = pyttsx3.init()

    # Obtener las voces disponibles
    voices = engine.getProperty('voices')

    # Establecer la voz deseada si se proporciona voice_id
    if voice_id is not None:
        for voice in voices:
            if voice.id == voice_id:
                engine.setProperty('voice', voice)
                break

    # Reproducir el texto
    engine.say(text)
    engine.runAndWait()
```
